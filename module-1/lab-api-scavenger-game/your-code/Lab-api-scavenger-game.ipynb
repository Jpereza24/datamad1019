```python
import os
from dotenv import load_dotenv
load_dotenv()
import requests
```

```python
def githubRequestAuthorized(resource):
    authToken = os.getenv("GITHUB_API_TOKEN")
    if not authToken:
        raise ValueError("NECESITAS UN TOKEN")
    else:
        print("We have a github token: ", authToken[0:4])
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
```

```python

scavenger = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents").json()

```

```python
scavenger
```

```python
path_lst = []
for x in scavenger:
    for key, value in x.items():
        if key == 'path':
            path_lst.append(x[key])
```

```python
path_list = path_lst[1:]
print(path_list)
```

```python
file_lst = []
for path in path_list:
    file_lst.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(path)).json())
    
```

```python
file_lst
```

```python
file_name = []
for x in file_lst:
    for y in x:
        for key, value in y.items():
            if key == 'name':
                file_name.append(value)
```
